# Pyplot API and Object-Oriented (OO) API
* Pylot API => Quickly 
* OO API => Advance 

### Pyplot API 

```python
x = np.linspace(0,10,1000)
```

```python
plt.plot(x,np.sin(x),color = 'pink',label = "sin(x)",linestyle = "dashed");
plt.plot(x,np.cos(x),color = 'blue',label = "cos(x)");
plt.title('Sin(x) and Cos(x)');
plt.xlabel('Bien X')
plt.ylabel('Sin(X)')
#plt.xlim([0,4]) #Gioi han do thi tu 0 den 4
#plt.ylim([-0.7,0,6 ])
#plt.axis([0,4,-0.7,0.6]) #Gioi han do thi theo truc tung(x) va truc hoanh(y)
plt.axis('equal') 
plt.legend();
```

### Object-Oriented (OO) API

```python
#1 Prepare Data
x = [1,2,3,4]
y= [11,22,33,44]
#2 Set up plot
fig,ax = plt.subplots(figsize=(5,5)) #Size 5x5 #Fig la hinh tong con ax la hinh nho
#3. Plot the data
ax.plot(x,y);
ax.set(title = 'Simple plot',xlabel = 'Truc X',ylabel = 'Truc Y');
```

### Scatter

```python
# Pylot API 
plt.scatter(x,np.exp(x)); #no.exp(x) = e^(x)
```

```python
## OO API 
# Prepare Data 
rng = np.random.RandomState(0)
x = rng.rand(100)
y = rng.rand(100)
size = 1000*rng.rand(100) 
colors = rng.rand(100)
fig,ax = plt.subplots()
im1 = ax.scatter(x,y,s = size,c = colors,cmap = 'viridis',alpha = 0.3); #Alpha la do trong suot,cmap la chinh bang mau
fig.colorbar(im1);
```

# Subplots

```python
# Option 1:
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot data
ax1.plot(x,x/2);
ax1.plot(x,x*2);
ax2.scatter(np.random.random(100),np.random.random(100),cmap = 'viridis',alpha = 0.3);
ax3.bar(soft_drinks.keys(),soft_drinks.values(),color = 'blue');
ax4.hist(student_height,color = 'blue');
```

```python
# Option 2:
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot data
ax[0,0].plot(x,x/2);
ax[0,0].plot(x,x*2);
ax[0,1].scatter(np.random.random(100),np.random.random(100),cmap = 'viridis',alpha = 0.3);
ax[1,0].bar(soft_drinks.keys(),soft_drinks.values(),color = 'blue');
ax[1,1].hist(student_height,color = 'blue');
```

# Project State California 

```python
cities = pd.read_csv('./california_cities.csv')
```

```python
cities.head()
#Find a Data Area (Before)
cities['area_total_km2'].sort_values(ascending= False)
```

```python
latitude,longitude = cities['latd'],cities['longd']
#Import more data
population, area = cities['population_total'],cities['area_total_km2']
latitude[:10],longitude[:10]
```

```python
# Pyplot API 
plt.style.use('seaborn')
plt.scatter(longitude,latitude,c= np.log10(population),cmap = 'viridis',
            s = area,linewidths=0,alpha = 0.5);
plt.axis('equal');
plt.xlabel('longitude');
plt.ylabel('latitude');
plt.colorbar(label = 'log$_(10)$(population)');
plt.clim(3,7)
plt.title('California Cities: Population and Area Distribution');
# Create A legend Size 
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,c ='k',alpha = 0.5,label= str(area) + 'km$^2$')
plt.legend(labelspacing=1,title = 'City Area');
```
